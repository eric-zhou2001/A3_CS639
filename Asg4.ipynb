```python
# load in the original dataset
df1 = pd.read_csv("archive/Datafiniti_Fast_Food_Restaurants.csv")
df1 = df1.drop_duplicates(subset=["address"])
df1
```

```python
# get a count of the occurrences of each restaurant
unique_count = df1.groupby(["name"])["name"].count().sort_values(ascending=False)
unique_count.head(10)
```

```python
# get only the top 10 most popular restaurants and convert that list to a set
rstrnts = [rstrnt for rstrnt in unique_count.head(10).keys()]
rstrnts = set(rstrnts)
rstrnts
```

```python
# only keep columns with physical information about the restaurants
# note - province = state
df1.columns
cols = ["latitude", "longitude", "address", "city", "country", "name", "postalCode", "province"]
cols
```

```python
# create a new df with only the specified columns
df2 = df1[cols]
df2
```

```python
# keep only the top 10 most popular restaurants
df3 = df2.loc[df2["name"].isin(rstrnts)]
df3 = df3.reset_index(drop=True)
df3 = df3.rename(columns={"province": "State Code"})
df3
```

```python
# create a new dataframe with info about states and the regions/divisions they belong too
regions_df = pd.read_csv("regions_divisions.csv")
regions_df = regions_df.rename(index={"State Code": "province"})
regions_df
```

```python
# label the regions of each fast food restaurant
final_df = df3.merge(regions_df, on=["State Code"])
final_df
```

```python
# convert 
final_df.to_csv("cleaned_rstrnts.csv")
```
